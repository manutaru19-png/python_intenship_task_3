```python
# loop_tasks.py

print("1. Printing numbers from 1 to 100 using for loop")
for i in range(1, 101):
    print(i, end=" ")
print("\n")

# --------------------------------------------------

print("2. Countdown timer using while loop")
count = 10
while count > 0:
    print("Countdown:", count)
    count -= 1
print("Time's up!\n")

# --------------------------------------------------

print("3. Demonstrating break and continue")
for i in range(1, 11):
    if i == 5:
        print("Skipping number 5 (continue)")
        continue
    if i == 9:
        print("Stopping loop at 9 (break)")
        break
    print(i)
print()

# --------------------------------------------------

print("4. Iterating over string characters")
name = "Python"
for char in name:
    print(char)
print()

# --------------------------------------------------

print("5. Multiplication table")
num = int(input("Enter a number for multiplication table: "))
for i in range(1, 11):
    print(f"{num} x {i} = {num * i}")
print()

# --------------------------------------------------

print("6. Using range with steps (even numbers from 2 to 20)")
for i in range(2, 21, 2):
    print(i, end=" ")
print("\n")

# --------------------------------------------------

print("7. Combining loop with conditions (even & odd)")
for i in range(1, 21):
    if i % 2 == 0:
        print(i, "is Even")
    else:
        print(i, "is Odd")
print()

# --------------------------------------------------

print("8. Real-world example: Exam pass/fail check")
marks = int(input("Enter your exam marks: "))

if marks < 0 or marks > 100:
    print("Invalid marks entered")
else:
    for _ in range(1):
        if marks >= 40:
            print("Result: Pass üéâ")
        else:
            print("Result: Fail ‚ùå")

print("\nProgram completed successfully.")
```
